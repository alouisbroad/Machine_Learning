```
'''
Function 1: Finding Proper Nouns
The following function takes a string and returns 
a string of proper nouns (people, places, organisations)
present in the original string.

By Alistair Broad
'''

def find_proper_nouns(text):
    try:
        import re                                                        # Used to remove all non-alpabetical symbols. 
        import nltk                                                      # Used in categorising text. 
        nltk.download("averaged_perceptron_tagger")
        text = re.sub('[^a-zA-Z]', ' ', text)                            # Removes all the non-alphabetical elements from the given text. 
        text_tag = nltk.tag.pos_tag(text.split())                        # NLTK .tag.pos_tag method categorieses each word noun, adjective etc. 
        prep_text = [w for w,t in text_tag if t == 'NNP' or t == 'NNPS'] # Iterates the dictonary removing entries that aren't proper nouns (nnp) or plural proper nouns (nnps).
        prep_text = [prop_nouns for prop_nouns in prep_text if len(prop_nouns) != 1] # Removes any words that are 1 letter long (Added as I found that "'s" left a loose "s" in the output).
        return(' '.join(prep_text))                                      # Joins and returns the result into single string. 

      # Throw an expection if the wrong type is given.
    except(ValueError):
        print("Incorrect value input. Ensure you have entered a string to the function.")

# Test for the function - a random extract from Beowulf. 
test_text = "Hrothgar answered, helm of the Scyldings: I remember this man as the merest of striplings. His father long dead now was Ecgtheow titled, Him Hrethel the Geatman granted at home his One only daughter; his battle-brave son Is come but now, sought a trustworthy friend. Seafaring sailors asserted it then,"
# Apply the function to the text. 
beowulf = find_proper_nouns(test_text)

# Prints successfully 
print(beowulf)

'''
Output:

[nltk_data] Downloading package averaged_perceptron_tagger to
[nltk_data]     /root/nltk_data...
[nltk_data]   Package averaged_perceptron_tagger is already up-to-
[nltk_data]       date!
Hrothgar Scyldings Ecgtheow Him Hrethel Geatman

'''
```

```
'''
Function 2: BBC Articles 
This Function is a web scraper function, it will enter 
user defined websites and return the date, title and
the main body of content. 

By Alistair Broad 
'''
# Three example bbc news articles.
url_example = 'https://www.bbc.co.uk/news/uk-52259844'
url_example2 = 'https://www.bbc.co.uk/news/uk-51004218'
url_example3 = 'https://www.bbc.co.uk/news/world-europe-52267376'

def bbc_articles(*urls):
    import requests
    from bs4 import BeautifulSoup
    joint_string = ''
      # Try this section. 
    try:
        for url in urls:                                      # Iterate over the url arguments given.
            response = requests.get(url)                        # Connect to the website.
            soup = BeautifulSoup(response.text, "html.parser")  # Download the webpage content. 
            joint_string += url + '\n'                          # Add the url to the output
            joint_string += soup.h1.text + "\n"                 # Add the title of the article. 
            joint_string += soup.find('div', class_ = 'date date--v2').text + "\n \n" # Adds the date to the string. 
            content = soup.find('div',attrs={"class":"story-body__inner"}) # Finds the main article content. 
            for p in content.select('p'):                       # Iterates through the "<p>" markers and adds them to the string. 
                joint_string += p.text + "\n"
            joint_string += "\n \n"                             # Add space at the bottom of the article before the next one. 
        return(joint_string)                                  # Return the combined string. 

  # Throw an expection if the wrong type is given.
    except(ValueError):
        print("Incorrect value input. Ensure all inputs are BBC news article URLs.")

# Printing the outputs of the 3 urls from above. 
print(bbc_articles(url_example, url_example2, url_example3))

# Applying the first function to the output of the second. 
find_proper_nouns(bbc_articles(url_example))

```
